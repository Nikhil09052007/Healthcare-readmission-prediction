# Diabetes Dataset - Medication Features Dictionary

This section provides detailed information about all **medication-related features** in the diabetes readmission dataset, prepared for Machine Learning engineers.

## 1. Individual Medications

These features are typically encoded as: **"No"**, **"Steady"**, **"Up"**, or **"Down"**.

| Medication            | Class                              | Description                                                                 | Key Relationships & Impact on Readmission |
|-----------------------|------------------------------------|-----------------------------------------------------------------------------|-------------------------------------------|
| **metformin**        | Biguanide                         | First-line oral antidiabetic drug for type 2 diabetes                      | Correlated with `diabetesMed`, `A1Cresult`. Stable use often linked to lower risk. |
| **repaglinide**      | Meglitinide (short-acting)        | Stimulates insulin release, used for post-meal glucose control             | Less common; interacts with `num_medications` |
| **nateglinide**      | Meglitinide                       | Similar to repaglinide                                                     | Low frequency |
| **chlorpropamide**   | 1st Generation Sulfonylurea       | Older sulfonylurea, higher risk of hypoglycemia                            | Rare in modern data |
| **glimepiride**      | 2nd Generation Sulfonylurea       | Common insulin secretagogue                                                | Common oral agent |
| **glipizide**        | 2nd Generation Sulfonylurea       | Very commonly used                                                         | Often used in combination |
| **glyburide**        | 2nd Generation Sulfonylurea       | Common sulfonylurea                                                        | Has combination form |
| **pioglitazone**     | Thiazolidinedione (TZD)           | Insulin sensitizer, affects lipids and weight                              | Cardiovascular considerations |
| **rosiglitazone**    | Thiazolidinedione (TZD)           | Similar to pioglitazone, restricted in many regions                        | Historical CV risk concerns |
| **acarbose**         | Alpha-glucosidase inhibitor       | Delays carbohydrate absorption                                             | GI side effects common |
| **miglitol**         | Alpha-glucosidase inhibitor       | Similar to acarbose                                                        | Rare |
| **insulin**          | Injectable hormone                | Indicates more severe diabetes or poor control                             | **Very strong predictor** of readmission |

## 2. Combination Medications

| Medication                  | Description                                      | Encoding                  | Notes |
|-----------------------------|--------------------------------------------------|---------------------------|-------|
| **glyburide-metformin**    | Fixed-dose combination of glyburide + metformin | No / Steady / Up / Down   | Indicates advanced therapy |
| **glipizide-metformin**    | Fixed-dose combination of glipizide + metformin | No / Steady / Up / Down   | Advanced oral therapy |

## 3. Related Medication Context Features

| Feature             | Type     | Description                                                      | Importance |
|---------------------|----------|------------------------------------------------------------------|----------|
| **change**          | Binary   | "Ch" = diabetes medication changed during admission, "No" = no change | High - strong predictor of readmission |
| **diabetesMed**     | Binary   | "Yes" = any diabetes medication was prescribed, "No" = none    | Very high |
| **num_medications** | Integer  | Total number of medications (diabetes + others) given during stay | High - proxy for comorbidity burden |

## Modeling Notes & Recommendations

- **Feature Engineering Ideas**:
  - Create aggregates: `num_diabetes_meds`, `has_insulin`, `has_sulfonylurea`, `has_TZD`
  - Interaction features: `insulin` × `A1Cresult`, `change` × `time_in_hospital`
  - Group rare medications to reduce sparsity

- **Key Insights**:
  - Insulin use and medication changes are among the **strongest predictors** of hospital readmission.
  - Patients on multiple diabetes medications (`num_medications` high) tend to have higher readmission rates.
  - Stable metformin-only regimens are often associated with better outcomes.

- **Encoding Tips**:
  - Use one-hot encoding or target encoding for medication states.
  - Consider ordinal encoding for "No" < "Steady" < "Up"/"Down" in some models.
